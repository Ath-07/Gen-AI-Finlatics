```python
# Import packages
from transformers import pipeline
import time
from IPython.display import clear_output
from datetime import datetime
from transformers import AutoTokenizer
```

```python
# Load the pre-trained Englishâ†’French model
translator = pipeline("translation_en_to_fr", model="Helsinki-NLP/opus-mt-en-fr")
tokenizer = AutoTokenizer.from_pretrained("Helsinki-NLP/opus-mt-en-fr")
print("âœ… Translation model loaded successfully!")
```

```python
#Translates English text into French using Hugging Face model.
def translate_to_french(text):
    translated = translator(text)
    return translated[0]['translation_text']
```

```python
#Character-by-character typing effect.
def typing_effect(text):
    for char in text:
        print(char, end="", flush=True)
        time.sleep(0.03)
    print()
```

```python
#Word-by-word streaming effect for chatbot messages.
def stream(text):
    print("French:", end=" ")
    words = text.split()
    for word in words:
        print(word, end=" ", flush=True)
        time.sleep(0.1)
    print()
```

```python
history = []
session_start = datetime.now()
input_tokens = 0
output_tokens = 0
```

```python
#Interactive English-to-French Chatbot.
def chatbot():
    global input_tokens, output_tokens
    clear_output(wait=True)
    print("ðŸ’¬ English-to-French Chatbot")
    print("Commands: history | clear | quit\n")

    while True:
        user_input = input("English: ")

        if user_input.lower() in ["quit", "exit"]:
            total_tokens = input_tokens + output_tokens
            print("French: Au revoir! ðŸ‘‹")
            print("\nðŸ”¢ Token Usage Summary")
            print("-" * 40)
            print("Input tokens :", input_tokens)
            print("Output tokens:", output_tokens)
            print("Total tokens :", total_tokens)
            print("-" * 40)
            print("French: Au revoir! ðŸ‘‹")
            break

        if user_input.lower() == "history":
            if not history:
                print("ðŸ“­ No history yet.")
            else:
                print("\nðŸ“œ Conversation History:")
                for i, item in enumerate(history, 1):
                    print(f"{i}. [{item['time']}] ðŸ‡¬ðŸ‡§ {item['english']}")
                    print(f"   [{item['time']}] ðŸ‡«ðŸ‡· {item['french']}")
            print("-" * 40)
            continue

        if user_input.lower() == "clear":
            history.clear()
            print("ðŸ§¹ History cleared.")
            print("-" * 40)
            continue

        input_tokens += len(tokenizer.encode(user_input))
        french_translation = translate_to_french(user_input)
        output_tokens += len(tokenizer.encode(french_translation))
        timestamp = time.strftime("%Y-%m-%d %H:%M:%S")
        history.append({
            "time": timestamp,
            "english": user_input,
            "french": french_translation,
            "input_tokens": input_tokens,
            "output_tokens": output_tokens
        })
        stream(french_translation)
        print()
```

```python
chatbot()
```
