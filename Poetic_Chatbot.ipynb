```python
#Importing requried liberaries of time aand google
import time
import random
from google import genai
from google.colab import userdata
```

```python
#Key to connect Gemini with API
key = userdata.get('Gemini_API')
if key is None:
    raise ValueError("Gemini API key not found. Please set it in Colab userdata.")
```

```python
poetic_mode = True
conversation_log = []
total_tokens = 0
TYPE_SPEED = 0.08

moods = ["romantic", "mystical", "hopeful", "playful", "melancholic"]
```

```python
#Adding stream effect for better effect on text
def stream(text):
  print('Bot:', end = ' ')
  words = text.split()
  for word in words:
    print(word, end = ' ', flush = True)
    time.sleep(0.1)
  print()
```

```python
def estimate_tokens(text):
    return int(len(text.split()) * 1.3)
```

```python
#Connect with Gemini
try:
  client = genai.Client(api_key=key)
except KeyError:
  print(f'Error with gemini api key.\n error message:\n{KeyError}')
```

```python
#giving poetic nature with prompt
def ask_gemini(user_input):
    global total_tokens

    mood = random.choice(moods)

    if poetic_mode:
        prompt = f"""
You are a poetic chatbot.
Respond in the following format:

Poem:
(2â€“3 short poetic lines in a {mood} tone)

Answer:
(Short and clear factual answer)

Question: {user_input}
"""
    else:
        prompt = f"Answer the following question clearly and concisely:\n{user_input}"

    try:
        total_tokens += estimate_tokens(prompt)

        response = client.models.generate_content(
            model="gemini-2.5-flash",
            contents=prompt
        )

        reply = response.text
        total_tokens += estimate_tokens(reply)

        return reply

    except Exception as e:
        return f"The muse falteredâ€”{str(e)}"
```

```python
#Use loops for chat
print("Welcome to your Poetic Gemini Chatbot!")
print("Commands: /poetry on | /poetry off | quit\n")

while True:
    user_input = input("You: ")

    if user_input.lower() == "quit":
        if conversation_log:
            summary_prompt = f"""
Summarize this conversation in two poetic lines:
{conversation_log}
"""
            summary = ask_gemini(summary_prompt)
            stream("Before you goâ€” " + summary)

        print(f"\nBot: Approximate tokens used: {total_tokens}")
        print("Bot: Farewell, till words meet again ðŸŒ™")
        break

    if user_input.lower() == "/poetry off":
        poetic_mode = False
        print("Bot: Poetic mode turned OFF.")
        continue

    if user_input.lower() == "/poetry on":
        poetic_mode = True
        print("Bot: Poetic mode turned ON ðŸŒ¼")
        continue

    conversation_log.append(user_input)
    response = ask_gemini(user_input)
    stream(response)
```
